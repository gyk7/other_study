```python
# 사용일자를 기준으로 사용월 column 만듦.
data['사용월'] = data['사용일자'].dt.month
```

```python
# 사용월을 int형으로 변환
data = data.astype({'사용월':'int'})
data
```

```python
# 계절 구분(사용일자를 기준으로)
def get_season(사용월):
    if 3 <= 사용월 <= 5:
        return '봄'
    elif 6 <= 사용월 <= 8:
        return '여름'
    elif 9 <= 사용월 <= 11:
        return '가을'
    else:
        return '겨울'
data['계절'] = data['사용월'].apply(get_season)
data
```

```python
# 노선번호를 기준으로 오름차순 정렬
sort_data = data.sort_values('노선번호')
sort_data
```

```python
# 노선번호 앞의 두글자가 10진수인지 아닌지 -> 마을버스 구분(마을버스는 앞의 두글자 지역명)
sort_data['is_decimal'] = sort_data['노선번호'].apply(lambda x: x[:2].isdecimal())

# 시내버스, 마을버스, 야간버스
city_bus = sort_data[sort_data['is_decimal']]
village_bus = sort_data[(~sort_data['is_decimal']) & (sort_data['노선번호'].str[0] !='N') ]
night_bus = sort_data[sort_data['노선번호'].str[0] =='N']
```
